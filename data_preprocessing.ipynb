```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("student_performance.csv")
df
```

```python
print(df.info())
print(df.describe(include=[np.number]).T)
```

```python
print(df.describe(include=[object]).T)
print("missing values per column:", df.isnull().sum())
```

```python
subject_cols = [
    "math_score", "science_score", "history_score", "geography_score"
]
df["overall_grade"] = df[subject_cols].sum(axis=1, skipna=True)
conditions = pd.DataFrame({
    "math_pass": df["math_score"] >= 50,
    "reading_pass": df["reading_score"] >= 50,
    "writing_pass": df["writing_score"] >= 50,
    "science_pass": df["science_score"] >= 50,
    "history_pass": df["history_score"] >= 50,
    "geography_pass": df["geography_score"] >= 50,
    "attendance_ok": df["attendance_rate"] >= 50
})
df["performance_score"] = conditions.sum(axis=1)
df.loc[df[subject_cols].isna().all(axis=1), "performance_score"] = np.nan
df
```

```python
df["age"] = pd.to_numeric(df["age"], errors="coerce")
subject_cols = [
    "math_score", "reading_score", "writing_score",
    "science_score", "history_score", "geography_score"
]
fill_zero_cols = ["attendance_rate"] + subject_cols
df["age"].fillna(df["age"].mean(), inplace=True)  
df[fill_zero_cols] = df[fill_zero_cols].fillna(0)  
print(df.isna().sum())
```

```python
df
```

```python
df['passed'] = np.where(
    (df['attendance_rate'] > 80) & 
    (df['math_score'] > 50) & 
    (df['science_score'] > 50) & 
    (df['history_score'] > 50) & 
    (df['geography_score'] > 50),
    'Yes', 'No'
)
df
```

```python
print(df['age'].max())
print(df['age'].min())
median_age = df['age'].median()
df['age'] = np.where((df['age'] < 10) | (df['age'] > 20), median_age, df['age'])
print(df['age'].max())
print(df['age'].min())
df['age'] = df['age'].astype(int)
```
